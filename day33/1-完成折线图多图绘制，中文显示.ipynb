```python
import matplotlib.pyplot as plt
import random
import matplotlib
from matplotlib import font_manager

from matplotlib.font_manager import FontProperties
```

```python
#  绘制x轴和y轴的刻度
x = range(2,26,2)
y = [random.randint(19,34) for i in x]

plt.figure(figsize=(20,8),dpi=80)
# 设置x,y的刻度
plt.xticks(x)
plt.xticks(range(1,25))
plt.yticks(y)
plt.yticks(range(min(y),max(y)+1))
# 设置x轴刻度标签
x_ticks_label =["{}:00".format(i) for i in x]
y_ticks_label = ["{}℃".format(i) for i in range(min(y),max(y)+1)]
# 字体旋转45°
plt.xticks(x, x_ticks_label, rotation=45)
plt.yticks(range(min(y),max(y)+1), y_ticks_label)
# 设置中文显示标题
plt.title('一天内许昌温度变化', fontproperties="STXinwei",fontsize=30)
plt.plot(x,y)
# plt.show()
plt.savefig('./t2.png')
```

```python
# 一图多线
# a和b每年的读书数量,x轴代表年份,y轴代表读书数量
y1 = [1, 5, 3, 12, 15, 4, 20]
y2 = [0, 0, 2, 5, 13, 3, 15]
x = range(2000,2007)
plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,y1,color='green', label='lily')
plt.plot(x,y2,color='red',label='Tom')
# 设置x和y刻度
xticks_labels = ['{}年'.format(i) for i in x]
my_font = font_manager.FontProperties(fname='C:\Windows\Fonts\STXINGKA.TTF',size=24)
plt.xticks(x,xticks_labels,rotation=45)

plt.legend(prop=my_font,loc='upper right')
# plt.show()
plt.savefig('./一图多线.png')
```

```python
# 绘制散点图
y = [35, 37, 16, 31, 32, 11, 12, 36, 36, 37, 38, 39, 32, 40, 34, 17, 18,
     31, 36, 17, 20, 14, 15, 15, 15, 19, 21, 22, 22, 23]
x= range(1,31)
# 设置图形大小
plt.figure(figsize=(20,8),dpi=80)
# 使用scatter绘制散点图
plt.scatter(x, y, label='6月')
xticks_labels = ['6月{}'.format(i) for i in x]
font_manager.FontProperties(fname='C:\Windows\Fonts\STXINGKA.TTF',size=24)
plt.xticks(x[::3],xticks_labels[::3],rotation=45)
plt.xlabel('日期', fontproperties=my_font)
plt.ylabel('温度', fontproperties=my_font)
# plt.show()
plt.savefig('./散点图.png')
```

```python
# 绘制条形图
# x轴代表电影名称，y轴表示票房数据
x = ['流浪地球','疯狂的外星人','飞驰人生','大黄蜂','熊出没·原始时代','新喜剧之王']
y = [38.13,19.85,14.89,11.36,6.47,5.93]
plt.figure(figsize=(16,9),dpi=160)
# 绘制条形图需设置条形的宽度width，
font_manager.FontProperties(fname='C:\Windows\Fonts\STXINGKA.TTF',size=24)
rects = plt.bar(range(len(y)), y, width=0.3, color='green')
plt.xticks(range(len(x)), x,fontproperties=my_font, rotation=45)
for rect in rects:
     height = rect.get_height()
     plt.text(rect.get_x() + rect.get_width()/2, height+0.3,str(height),ha='center')
# plt.show()
plt.savefig('./条形图.png')
```
