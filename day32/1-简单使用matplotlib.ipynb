```python
# 导入模块
import matplotlib.pyplot as plt
# 传入x和y，通过plot画图
plt.plot([1, 0, 9, 7], [4, 5, 6, 5.5])
# 执行程序展示图形
plt.show()
```

```python
from matplotlib import pyplot as plt
x = range(1, 10)
y = [7, 8, 9, 15, 18, 19, 26, 30, 36]
plt.plot(x, y, color='red',alpha=0.5,linestyle='--',linewidth=3)
plt.show()
```

```python
# 折点样式
from matplotlib import pyplot as plt
x = range(1,8) # x轴的范围
y = range(31, 38)  # y轴的范围
plt.plot(x, y, marker='o')
plt.show()

```
